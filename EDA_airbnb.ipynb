```python
import pymongo
import pandas as pd
import numpy as np
```

```python
client = pymongo.MongoClient("mongodb://localhost:27017")
db = client["sample_airbnb"]
collention = db["sample"]
df = pd.DataFrame(list(collention.find({})))
```

```python
df.head(20).T
```

```python
df2 = df[['_id','name','description','price','amenities','bedrooms','beds','cleaning_fee','guests_included','maximum_nights','minimum_nights','number_of_reviews','property_type','room_type','availability','review_scores']]
```

```python
df2.head().T
```

```python
df_host = df['host'].apply(pd.Series)
host_df1 = df_host[['host_id','host_name']]

df_address = df['address'].apply(pd.Series)

df_avail = df2['availability'].apply(pd.Series)

df3 = pd.concat([df2,host_df1,df_address,df_avail],axis=1)
```

```python
df3.head().T
```

```python
df3.info()
```

```python
df3['review_scores'] = df3['review_scores'].apply(lambda x: x['review_scores_rating'] if isinstance(x,dict) and 'review_scores_rating' in x else "NaN")
```

```python
df3['_id'] = df3['_id'].astype(str).astype(int)
df3['price'] = df3['price'].astype(str).astype(float)
df3['cleaning_fee'] = df3['cleaning_fee'].astype(str).astype(float)
df3['guests_included'] = df3['guests_included'].astype(str).astype(int)
df3['maximum_nights'] = df3['maximum_nights'].astype(str).astype(int)
df3['minimum_nights'] = df3['minimum_nights'].astype(str).astype(int)
df3['review_scores'] = df3['review_scores'].astype(str).astype(float)
```

```python
df3['host_id'] = df3['host_id'].astype(str).astype(int)
```

```python
df3['bedrooms'] = df3['bedrooms'].replace(0, np.nan)

df3['beds'] = df3['beds'].replace(0, np.nan)
```
